```python
#Imports libraries
import tensorflow as tf
import yfinance as yf
import pandas as pd
import numpy as np

import requests
import random
import json
import io

import matplotlib.pyplot as plt
from tqdm.auto import tqdm
```

```python
#Get companies ticker
url="https://dumbstockapi.com/stock?format=tickers-only&exchanges=NYSE,NASDAQ,AMEX"

url_content = requests.get(url).content
tickers = json.loads(url_content)
random.shuffle(tickers) # Shuffle
tickers[:5] # Examples
```

```python
num_companies = 100
chart_items = ['Close', 'Volume']
chart_data_raw = yf.download(
        tickers = random.sample(tickers, num_companies),
        actions=True,
        period = "max",
        interval = "1d",
        group_by = 'ticker',
        prepost = True,
        threads = True,
        proxy = None
    ).drop(columns=yf.shared._ERRORS.keys()) # Drop errors
print(num_companies-len(yf.shared._ERRORS.keys()), "Successful download")
chart_data = chart_data_raw.loc[:, pd.IndexSlice[:, chart_items]]
chart_data.sample(axis=1).plot()
```

```python
#Try tpu strategy
try:
  resolver = tf.distribute.cluster_resolver.TPUClusterResolver(tpu='')
  tf.config.experimental_connect_to_cluster(resolver)
  tf.tpu.experimental.initialize_tpu_system(resolver)
  
  strategy = tf.distribute.TPUStrategy(resolver)
  print("All devices: ", tf.config.list_logical_devices('TPU'))

except:
  strategy = tf.distribute.get_strategy() # Normal strategy
  print("All devices: ", tf.config.list_logical_devices())
```

```python
#Split data
chart_size = 100
features = {
  "chart_data": [],
  "chart_span": [],
  "target_distance": [],
}
targets = {
  "next_close": [],
}

def days_between(start, end):
  return (end - start) / np.timedelta64(1, 'D')

for symbol, company in tqdm(chart_data.groupby(level=0, axis=1)):
  with pd.option_context('mode.use_inf_as_null', True):
    company_dropna = company.dropna() # Drop nulls
  value = company_dropna.values
  company_index = company_dropna.index

  for i in range(len(value) - chart_size):
    chart_indices = slice(i, chart_size + i)
    chart = value[chart_indices]

    target_index = chart_size + i
    target = value[target_index]

    if (target > 0).all() and (chart > 0).all():
      targets["next_close"].append(target[0])
      features["chart_data"].append(chart)
      
      chart_dates = company_index[chart_indices]
      target_date = company_index[target_index]

      days = days_between(chart_dates[0], chart_dates[-1])
      target_days = days_between(chart_dates[-1], target_date)
      features["chart_span"].append([days])
      features["target_distance"].append([target_days])

print()

#Convert dict items to tensor
def dict_items_to_tensor(dictionary):
  for key in dictionary:
    if tf.is_tensor(dictionary[key]):
      print(key, "is already tensor")
      continue
    dictionary[key] = tf.convert_to_tensor(dictionary[key])
    print(key, tf.shape(dictionary[key]))
dict_items_to_tensor(features)
dict_items_to_tensor(targets)

print()

#Get inputs and normalizers 
def numeric_normalization(name, features, axis=None):
  input = tf.keras.Input(shape=features[name].shape[1:], name=name)

  normalizer = tf.keras.layers.Normalization(axis, name=name+"_norm")
  normalizer.adapt(features[name])

  print(name, "Mean:", normalizer.mean)
  return input, normalizer

#Features
inputs = []

features['chart_data'] = tf.math.log(features['chart_data']) # Log normalize
chart_input, chart_norm = numeric_normalization('chart_data', features, 2)
inputs.append(chart_input)

single_inputs = []
single_norms = []
single_features = ['chart_span', 'target_distance']
for name in single_features:
  input, norm = numeric_normalization(name, features)
  single_norms.append(norm)
  single_inputs.append(input)
  inputs.append(input)

print()

#Normalize
features['chart_data'] = chart_norm(features['chart_data'])
tf.debugging.check_numerics(features['chart_data'], message='chart_data')
for i, name in enumerate(single_features):
  features[name] = single_norms[i](features[name])
  tf.debugging.check_numerics(features[name], message=name)
```

```python
#Plot data
fig, (history_inputs_plot, numeric_inputs_plot, target_distance_plot, targets_plot) = plt.subplots(4)

history_inputs_plot.plot(random.choice(features["chart_data"]))
history_inputs_plot.set_title('Random chart_data', loc='right')

numeric_inputs_plot.plot(features["chart_span"])
numeric_inputs_plot.set_title('All chart_span', loc='right')

target_distance_plot.plot(features["target_distance"])
target_distance_plot.set_title('All target_distance', loc='right')

targets_plot.plot(targets["next_close"])
targets_plot.set_title('All next_close', loc='right')
```

```python
#Create model
from tensorflow.keras import layers

x = layers.Conv1D(32, 3, activation='swish')(chart_input)
x = layers.MaxPool1D(2)(x)
x = layers.Conv1D(64, 3, activation='swish')(x)
x = layers.MaxPool1D(2)(x)
x = layers.Conv1D(64, 3, activation='swish')(x)
x = layers.MaxPool1D(2)(x)
x = layers.Conv1D(64, 3, activation='swish')(x)
x = layers.MaxPool1D(2)(x)
x = layers.Conv1D(64, 3, activation='swish')(x)
x = layers.MaxPool1D(2)(x)
x = layers.Flatten()(x)
x = layers.concatenate([x, *single_inputs])
x = layers.Dense(32, activation='swish')(x)
x = layers.Dense(16, activation='swish')(x)
x = layers.Dense(8, activation='swish')(x)
x = layers.Dense(4, activation='swish')(x)
x = layers.Dense(2, activation='swish')(x)
x = layers.Dense(1, name='next_close')(x)
with strategy.scope():
  model = tf.keras.Model(inputs, x, name='Stock_AI')
  model.compile(optimizer='adam',
                  loss='mean_squared_error',
                  metrics=['mean_absolute_error'],
                  steps_per_execution=None)
model.summary()
```

```python
#Plot model
tf.keras.utils.plot_model(model, show_shapes=True, rankdir="LR")
```

```python
#Train
history = model.fit(features, targets, epochs=5, validation_split=0.2, batch_size=None, callbacks=[cp_callback])
```

```python
def plot_history(history):
  plt.plot(history.history['loss'], label='loss')
  plt.plot(history.history['val_loss'], label='val_loss')
  plt.ylabel('Error')
  plt.xlabel('Epoch')
  plt.legend()
  plt.grid(True)
  
plot_history(history)
```
